```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1 = [0.1,0.2,0.3,0.4]
y1 = [0.5,2.2,3.2,8.7]
x2 = [0.1,0.2,0.3,0.4]
y2 = [1,4,9,16]
```

```python
plt.plot(x1,y1, label='first plot')
plt.plot(x2,y2, label='second plot')
plt.legend()
```

# The NumPy Library
Short for numerical python

includes a data structure called an ndarray (n-dimensional array, Numpy,..)

```python
num = 100
x = np.linspace(0, 20, num)
print(x)
```

```python
y = np.random.rand(num)
print(y)
```

```python
plt.plot(x,y)
```
